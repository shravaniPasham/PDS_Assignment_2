```python
import pandas as pd
```

```python
data=pd.read_csv("/content/train.csv")

```

```python
data.head()
```

```python
#checking for null values
data.isna().sum()
```

```python
data1 = pd.DataFrame(data)
```

```python
#(a)Replacing the null values of Mileage, Engine, Power, Seats and New_Price
#Median is used to avoid outliers for Mileage, Engine and Power
#Mode is used for Seats, as seats are only two types(5 or 7)
# Dropping the New_Price column, because there are two many null values
```

```python
# To perform the above operations, we need to convert the float values to int values

###(b)
#  Removing the units from Mileage, Engine, Power and New_Price
data1['Mileage'] = data1['Mileage'].str.extract('(\d+\.\d+)').astype(float)

data1['Engine'] = data1['Engine'].str.replace(' CC', '').astype(float)

data1['Power'] = data1['Power'].str.extract('(\d+\.\d+)').astype(float)

data1['New_Price'] = data1['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
data1['Mileage'].fillna(data1['Mileage'].median(),inplace=True)
data1['Engine'].fillna(data1['Engine'].median(),inplace=True)
data1['Power'].fillna(data1['Power'].median(),inplace=True)
data1['Seats'].fillna(data1['Seats'].mode()[0],inplace=True)

data1.drop('New_Price', axis=1, inplace=True)
```

```python
data1.head()
```

```python
output_csv_file = '/content/cleaned_data.csv'
data1.to_csv(output_csv_file, index = False)
```

```python
data2 = pd.read_csv('/content/cleaned_data.csv')
```

```python
data2.isna().sum()
```

```python
# (c)
# Changing the "Fuel_Type" and "Transmission" into numerical values

data2['Fuel_Type'].unique()
data2['Transmission'].unique()
data2['Owner_Type'].unique()

data2['Fuel_Type'].replace({'Diesel': 0, 'Petrol': 1, 'Electric': 2}, inplace=True)
data2['Transmission'].replace({'Manual': 0, 'Automatic': 1}, inplace=True)
data2['Owner_Type'].replace({'First': 1, 'Second': 2, 'Third': 3, 'Fourth & Above': 4}, inplace=True)

output_csv_file = '/content/encodedData.csv'
data2.to_csv(output_csv_file, index = False)

data2.head()
```

```python
#(d)
# Adding one more column called car_type where if the car has 5seats it is a sedan and if it has 7seats then it is a SUV.

def car_type(Seats):
  if Seats == 5:
    return "Sedan"
  else:
    return "SUV"

data2['Car_Type'] = data2['Seats'].apply(car_type)

output_csv_file = '/content/mutatedData.csv'
data2.to_csv(output_csv_file, index = False)
data2.head()
```
